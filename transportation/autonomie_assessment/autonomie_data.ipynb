```python
import pandas as pd
```

```python
ldv = pd.read_excel("ANL - TAPS-2303 Report  - Light Duty - Autonomie Results.xlsx", sheet_name='Autonomie results')

ldv_cols = [
    'Vehicle Powertrain: {string}',
    'Model Years: {years}',
    'Technology Progress: {string}',
    'Vehicle Class: {string}',
    'Vehicle Performance Category: {string}',

    'Battery Total Energy Beginning of Life: {Wh}',
    'Adjusted CD Range (Rdca) on Combined: {mile}',

    'Glider Manufacturing Cost {2023$}',
    '12V Battery Manufacturing Cost {2023$}',
    'High Voltage Battery Charger Manufacturing Cost {2023$}',
    'High Voltage Battery Manufacturing Cost {2020}',   # Column descriptions shown in 2023 USD, data table shown in 2020 USD
    # 'Battery Manufacturing Cost {2023$}',
    'Motor 1 Manufacturing Cost {2023$}',
    'Motor 2 Manufacturing Cost {2023$}',
    'Torque Converter Cost {2023$}',
    'Engine and After-Treatment Manufacturing Cost - Total {2023$}',
    'Transmission Manufacturing Cost - Total {2023$}',
    'Electrical Accessory Manufacturing Cost {2023$}',
    'Mechanical Accessory Manufacturing Cost {2023$}',
    'Starter Manufacturing Cost {2023$}',
    'Generator Manufacturing Cost {2023$}',
    'Fuel Cell System Cost {2020$)',                    # Column descriptions shown in 2023 USD, data table shown in 2020 USD
    'Hydrogen Storage Cost {2023$}',
    'DC/DC converter cost {2023$}',
    'Boost Converter Manufacturing Cost {2023$}',
    'Wheel Cost {2023$}',
    'Vehicle Propulsion Architecture Cost {2023$}',
    'Vehicle Manufacturing Cost {2023$}',

    'Adjusted Fuel Economy, Combined 55/45 - sticker, CS, Fuel (Gas.Equivalent): {mile/gallon}',
    'Utility-weighted Adjusted PHEV Fuel Economy on Combined, (Gas. Equivalent): {mile/gallon}',
    'Utility-weighted Adjusted PHEV Electrical Consumption on Combined: {W.h/mile}',
    'Adjusted Electricity Consumption, Combined 55/45 - sticker, CD (Wh/mi) With Charger'
]

ldv = ldv[ldv_cols]
ldv.fillna(0, inplace=True)
ldv.head()
```

```python
powertrain_list = [
    'BEV150',
    'BEV200',
    'BEV300',
    'BEV400',
    'Conventional CI',
    'Conventional CNG',
    'Conventional SI',
    'EREVPHEV 35AER',   # Newer voltec technology is not a series hybrid, it also has two motors and generators
    'EREVPHEV 50AER',
    'FCEV',
    # 'Par HEV CI',
    # 'Par HEV SI',
    # 'Par HEV SI Turbo',
    # 'Par PHEV35', # Older technology, only one motor and generator
    # 'Par PHEV50',
    'SplitHEV',     # Power-split hybrids are more adopted (e.g., Toyota Prius)
    # 'SplitPHEV'   # Only available as 20-mile AER
]

ldv = ldv[ldv['Vehicle Performance Category: {string}'] == 'Base']
ldv = ldv[ldv['Vehicle Powertrain: {string}'].isin(powertrain_list)]
ldv = ldv[ldv['Model Years: {years}'].isin([2023, 2025, 2030, 2035, 2050])]
ldv = ldv[ldv['Technology Progress: {string}'] == 'Low']
```
